#Reprogramando el envejecimiento: la ingeniería biológica detrás de una longevidad saludable"

##Tres ideas importantes

```
Envejecer es como acumular errores en un software.
Nuestro cuerpo funciona por medio de instrucciones (ADN).
Con el pasar de los años, el sistema (el cuerpo) va acumulando errores que van dañando las partes del mismo como nos dañaría a nosotros los musculos, la piel, la memoria y el corazón.
```

```
El cuerpo tiene mecanismos de defensa, pero se desgastan.
El cuero humano tiene varios anticuerpos que son nuestras defensas como lo son los globulos blancos, enzimas, etc.
Con el paso de la edad estas se van desgastando y perdiendo toda su fuerza lo que hace que muchas veces con una edad mayor seamos más propensos a enfermedades.
```

```
Podemos influir en cómo envejecemos con hábitos diarios.
Tal como un sistema tecnológico no podemos detener el tiempo para que no pasen los años pero, si podemos "programar" nuestro cuerpo con habitos diarios y saludables como por ejemplo, buena alimentación, dormir bien y hacer ejercicio para tener un mejor rendimiento en el futuro.
```
